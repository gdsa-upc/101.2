```python
# Leemos las librerias 
import numpy as np
import cv2
from drawmatches import drawMatches
```

```python
# Leemos las dos imágenes las cuales vamos a comparar que son de la misma clase
img = cv2.imread('C:/Users/papa/Desktop/Entregas/ajuntament_36.jpg')
img2 = cv2.imread('C:/Users/papa/Desktop/Entregas/ajuntament_37.jpg')
```

```python
# Iniciamos el detector SIFT
sift = cv2.SIFT()
```

```python
# Detectamos puntos de interés, los computamos y los ubicamos
kp, des = sift.detectAndCompute(img,None)
kp1, des1 = sift.detectAndCompute(img2,None)
```

```python
# Utilizamos las siguientes funciones para poder comparar los puntos de interés de las imágenes
bf = cv2.BFMatcher()
matches = bf.match(des,des1)
matches = sorted(matches,key=lambda val: val.distance)
```

```python
# Dibujamos las líneas que comparan los puntos de las dos imágenes
comparacion = drawMatches(cv2.cvtColor(img,cv2.COLOR_BGR2GRAY), kp, cv2.cvtColor(img2,cv2.COLOR_BGR2GRAY), kp1, matches[:300])
```

```python
# Y por último hacemos un imshow para ver el resultado
imshow(comparacion)
show()
```
